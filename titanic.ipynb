```python
import pandas as pd
import numpy as np
data = pd.read_csv('titanic.csv')
data['Age'].fillna(data['Age'].median(), inplace=True)
data['Embarked'].fillna(data['Embarked'].mode()[0], inplace=True)
data['Fare'].fillna(data['Fare'].median(), inplace=True)
data.drop(columns=['Cabin'], inplace=True)
data['Sex'] = data['Sex'].map({'male': 0, 'female': 1})
data['Embarked'] = data['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})
data['FamilySize'] = data['SibSp'] + data['Parch'] + 1
data.drop(columns=['PassengerId', 'Name', 'Ticket'], inplace=True)
print(data.head())
```

## FINDING ANY OUTLAYERS

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.boxplot(y='Fare', data=data)
plt.title('Box Plot of Fare')
plt.ylabel('Fare')
plt.show()
```

## REMOVING OUTLAYERS

```python
import numpy as np

Q1 = data['Fare'].quantile(0.25)
Q3 = data['Fare'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f"Lower Bound: {lower_bound}")
print(f"Upper Bound: {upper_bound}")

median_value = data['Fare'].median()
print(f"Median value: {median_value}")

data['Fare'] = np.where((data['Fare'] < lower_bound) | (data['Fare'] > upper_bound), median_value, data['Fare'])

print(data.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.boxplot(y='Fare', data=data)
plt.title('Box Plot of Fare')
plt.ylabel('Fare')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


age_bins = [0, 18, 35, 50, 65, 100]
age_labels = ['0-18', '19-35', '36-50', '51-65', '66+']
titanic['AgeGroup'] = pd.cut(titanic['Age'], bins=age_bins, labels=age_labels, right=False)


age_group_counts = titanic['AgeGroup'].value_counts().sort_index()


plt.figure(figsize=(8, 8))
plt.pie(age_group_counts, labels=age_group_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Age Groups')
plt.show()
```

```python
from scipy.stats import chi2_contingency

file_path = "titanic.csv"
data = pd.read_csv(file_path)

contingency_table = pd.crosstab(data["Fare"],data["Pclass"])

chi2_result = chi2_contingency(contingency_table,correction = False)
print("chi square result:",chi2_result)
```

```python
p_value = 6.562537418561338e-39
alpha = 0.05
if p_value<alpha:
    print("Reject the null hypothesis. There is a link")
else:
    print("Fail to reject the null hypothesis")
```
